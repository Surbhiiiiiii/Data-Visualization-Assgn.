```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("C:/Users/hp/Downloads/bank_csv.csv")
data
```

```python
#simple EDA
data.columns
data.head()
data.tail()
data.info()
data.describe()
data.corr()
sns.heatmap(data.corr(),annot=True)
```

```python
#univariate and Bivariate analysis
#univariate analysis:-histogram and boxplot
data.columns
sns.displot(data['age'])
```

```python
#kde:-Kernel density entity
sns.displot(data['age'],kde=True,bins=10)
sns.displot(data['age'],kde=False,bins=10)
```

```python
#boxplot
sns.boxplot(x=data['age'])
```

```python
#When input and Output is continuous:-Scatter plot
#if both the parameters aret categorical in nature:-bivariate analysis
#one is categorical and other is numerical:-barplot,boxplot
#univariate analysis  of  categorical variable: countplot
```

```python
#bivariate analysis:-
plt.figure(figsize=(10,6))
plt.scatter(data['marital'],data['balance'])
plt.title("Marital status vs. Balance")
plt.xlabel("Marital Status")
plt.ylabel("Balance")
plt.show()
```
